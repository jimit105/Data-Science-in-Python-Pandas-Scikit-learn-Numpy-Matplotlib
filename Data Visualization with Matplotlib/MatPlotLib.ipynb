## Basic Plotting, Plotting Terminologies

```python
x = np.linspace(0, 10, 100) #100 Data equally spaced between 0 to 10
y = np.sin(x)
z = np.cos(x)
```

```python
x
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])
#Params- left, buttom, width, height

ax.plot(x,y)
ax.plot(x,z)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])
#Params- left, buttom, width, height

#Change color of graph
ax.plot(x,y,c='red')
ax.plot(x,z,c='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])
#Params- left, buttom, width, height

#Transparency
ax.plot(x,y,c='red',alpha=0.4)
ax.plot(x,z,c='green',alpha=0.8)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])
#Params- left, buttom, width, height

#Highlight point/markers
ax.plot(x,y,c='red', alpha=0.8, marker='^')
ax.plot(x,z,c='green', alpha=0.8, marker='x')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])
#Params- left, buttom, width, height

#Change line width
ax.plot(x,y,c='red', alpha=0.8, lw=2.5)
ax.plot(x,z,c='green', alpha=0.8)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])
#Params- left, buttom, width, height

ax.plot(x,y,c='red', alpha=0.8, marker='^', lw=2.5)
ax.plot(x,z,c='green', alpha=0.8, marker='x')

#Change limit of axes
ax.set_xlim(0,10)
ax.set_ylim(-5,5)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 1, 1])
#Params- left, buttom, width, height

ax.plot(x,y,c='red', alpha=0.8, marker='^', lw=2.5, label='sine')
ax.plot(x,z,c='green', alpha=0.8, marker='x', label='cosine')

#Change limit of axes
ax.set_xlim(0,10)
ax.set_ylim(-1,1)

#Legend
ax.legend()

#Axes title
ax.set_xlabel('Time')
ax.set_ylabel('Amp')
ax.set_title('Sine and Cosine graph using Matplotlib')
```

## Subplots

```python
plt.subplots(1,2)
# Params- number of sub-graphs in row*column
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
#type(ax)

# ax[0] --> corresponds to first graph
# ax[1] --> corresponds to second graph

ax[0].plot(x,y,c='red')
ax[1].plot(x,z,c='green')
```

```python
#Change figure size
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,7))

ax[0].plot(x,y,c='red')
ax[1].plot(x,z,c='green')
```
